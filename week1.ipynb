```python
### Week 1: Basics of Artificial Intelligence

# Task 1: Install Python and Set Up AI Libraries

# already got it installed so nothing needs to be done


# Task 2: Write a Python Script with NumPy

import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

addition = A + B
subtraction = A - B
multiplication = A * B  # Element-wise multiplication
dot_product = np.dot(A, B)

# Print results
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nAddition:\n", addition)
print("\nSubtraction:\n", subtraction)
print("\nElement-wise Multiplication:\n", multiplication)
print("\nDot Product:\n", dot_product)
```

```python
# Task 3: Create and Manipulate a Data Table with pandas

import pandas as pd


data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math': [85, 90, 78, 88],
    'Science': [92, 87, 85, 95]
}

df = pd.DataFrame(data)

df['Total'] = df['Math'] + df['Science']

mean_scores = df.mean(numeric_only=True)
median_scores = df.median(numeric_only=True)

filtered_df = df[df['Math'] > 80]

print("\nDataset:")
print(df)
print("\nMean Scores:")
print(mean_scores)
print("\nMedian Scores:")
print(median_scores)
print("\nFiltered Students (Math > 80):")
print(filtered_df)
```
